```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# ==========================================
# i) CODE IMPLEMENTATION (Instruction i)
# ==========================================
# Load the image and convert it to grayscale
image = cv2.imread('Fish.jpg')  

if image is None:
    # Fallback placeholder if Fish.jpg isn't in your folder yet
    image = np.zeros((400, 400, 3), dtype=np.uint8)
    cv2.circle(image, (200, 200), 80, (255, 255, 255), -1)
    cv2.rectangle(image, (50, 50), (150, 150), (255, 255, 255), -1)

gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Apply Gaussian Blur to smooth out pixel noise
blurred = cv2.GaussianBlur(gray, (5, 5), 0)

# Run Canny Edge Detection (Using your repository baseline values)
canny_edges = cv2.Canny(blurred, 50, 150)

# Display the original and the edge-detected result side-by-side
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title("Original Image")
plt.axis("off")

plt.subplot(1, 2, 2)
plt.imshow(canny_edges, cmap='gray')
plt.title("Canny Edge Detection")
plt.axis("off")

plt.tight_layout()
plt.show()
```

```python
==================================================
ii) DISCUSSION OF DETECTED EDGES
==================================================
The Canny algorithm successfully isolates the structural boundaries of the subject.
By using non-maximum suppression, it eliminates thick gradient blurs and outputs
sharp, single-pixel-wide white lines tracking the key profile contours and structural
details against a clean black background.

    

==================================================
iii) IMPACT OF PARAMETER SETTINGS (HYSTERESIS THRESHOLDS)
==================================================
* Upper Threshold (T2 = 150): Sets the strict baseline for strong edges. Any gradient
  change sharper than 150 is permanently preserved. If this is turned too high, faint
  but genuine structural lines will vanish from the edge map.
* Lower Threshold (T1 = 50): Sets the filter boundary for weak edges. Gradients below
  50 are discarded entirely. Gradients between 50 and 150 are only kept if they physically
  connect directly to a strong edge. Setting this too low allows random camera textures
  and background noise to show up as broken, faulty edge fragments.
```
